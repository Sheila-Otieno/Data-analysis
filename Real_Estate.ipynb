```python
#import numpy and pandass libraries
import pandas as pd
import numpy as np
```

```python
#read csv file
data = pd.read_csv('Real_Estate_Sales_2001-2020_GL.csv')
```

```python
#check for data types
data.dtypes
```

```python
#change type of year from object to datateime
data['List Year'] = pd.to_datetime(data['List Year'], format = '%Y')
```

```python
data.dtypes
```

```python
#change type of date recorded from object to datateime
data['Date Recorded'] = pd.to_datetime(data['Date Recorded'], errors="coerce")
```
