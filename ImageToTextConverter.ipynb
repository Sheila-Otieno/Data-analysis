```python
import os
import pytesseract
from PIL import Image
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# If Tesseract is not in your PATH, specify the full path to the executable
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

def convert_images_to_text(image_folder):
    # List all files in the specified directory
    image_files = [f for f in os.listdir(image_folder) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

    # Dictionary to hold the filename and corresponding text
    text_output = {}

    for image_file in image_files:
        try:
            # Construct the full file path
            file_path = os.path.join(image_folder, image_file)
            # Open the image
            img = Image.open(file_path)
            # Use pytesseract to do OCR on the image
            text = pytesseract.image_to_string(img)
            text_output[image_file] = text
            print(f"Processed {image_file} successfully.")
        except Exception as e:
            print(f"Error processing {image_file}: {e}")

    return text_output

def save_text_to_pdf(text_dict, output_pdf):
    # Create a PDF canvas
    c = canvas.Canvas(output_pdf, pagesize=letter)
    width, height = letter

    for image_file, text in text_dict.items():
        # Add the filename as a title
        c.setFont("Helvetica-Bold", 14)
        c.drawString(72, height - 72, f"Text from {image_file}:")
        c.setFont("Helvetica", 12)

        # Set the starting position for the text
        y = height - 100

        # Split the text into lines and add to the PDF
        for line in text.splitlines():
            if y < 72:  # Avoid going out of margins
                c.showPage()  # Create a new page
                c.setFont("Helvetica-Bold", 14)
                c.drawString(72, height - 72, f"Text from {image_file}:")
                c.setFont("Helvetica", 12)
                y = height - 100  # Reset position

            c.drawString(72, y, line)
            y -= 15  # Move down for the next line

        c.showPage()  # Start a new page for the next image

    c.save()  # Save the PDF

# Specify the folder containing images
image_folder = '/content/images'  # Update this path
output_pdf = 'output_text.pdf'  # The output PDF file name

# Convert images to text
extracted_text = convert_images_to_text(image_folder)

# Save the extracted text to a PDF file
save_text_to_pdf(extracted_text, output_pdf)

print(f"Text extracted and saved to {output_pdf}.")
```
