```python
data = pd.read_csv('Measurements Dataset.csv')
data.head()
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
data['downtime%'] = (data['downtime']/ data['downtime'].sum())*100
data.head(50)
```

```python
data.sort_values('component', inplace=True)
```

```python
data["risk"] = data["component"].rank(method ='average')
```

```python
data.head(50)
```

```python
data['component'].unique()
```

```python
# prompt: draw a graph showing each components downtime and risk

import matplotlib.pyplot as plt

# Group data by component and sum downtime
component_downtime = data.groupby('component')['downtime'].sum()

# Create a bar chart showing downtime for each component
plt.figure(figsize=(12, 6))
plt.bar(component_downtime.index, component_downtime.values)
plt.xlabel('Component')
plt.ylabel('Downtime')
plt.title('Downtime for Each Component')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()

# Create a scatter plot showing downtime vs. risk for each component
plt.figure(figsize=(12, 6))
plt.scatter(data['component'], data['downtime'], c=data['risk'], cmap='viridis')
plt.xlabel('Component')
plt.ylabel('Downtime')
plt.title('Downtime vs. Risk for Each Component')
plt.xticks(rotation=90)
plt.colorbar(label='Risk')
plt.show()
```

```python
component_downtime = data.groupby('component')['risk'].sum()

# Create a bar chart showing downtime for each component
plt.figure(figsize=(12, 6))
plt.bar(component_downtime.index, component_downtime.values)
plt.xlabel('Component')
plt.ylabel('Risk')
plt.title('Risk for Each Component')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
data['Component_risk'] = data.groupby('component')['downtime%'].rank(ascending=True)
```
