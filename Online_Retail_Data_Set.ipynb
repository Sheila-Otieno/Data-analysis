```python
data = pd.read_excel('Online Retail Data Set.xlsx')
data.head()
```

```python
data.info()
```

```python
data['Invoice_Date'] = pd.to_datetime(data['InvoiceDate']).dt.date
data['InvoiceTime'] = pd.to_datetime(data['InvoiceDate']).dt.time
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data = data.astype({'CustomerID': 'int'})
```

```python
data.to_csv("Online_Retail_Data_Set_clean.csv")
```
